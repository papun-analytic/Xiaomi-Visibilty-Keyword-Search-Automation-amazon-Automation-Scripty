```python
import pandas as pd
import numpy as np
import os
import glob
import math
s1=input("give the brand input file")
s2=input("Give the csv file as input which have the list of sku whose brand mapping is required with column is SKU")
b=pd.read_csv(s1)
s=pd.read_csv(s2)
s["Brand"]=""
for brand in b["Brand"].str.lower():
    for i,sku in enumerate(s["SKU"].str.lower()):
        if brand in sku:
            s.loc[i,"Brand"]=brand
s["Brand"]=s["Brand"].str.title()
s["SKU"]=s["SKU"].str.title()
s["Source"]="Amazon"
s["Class"]="Keyword Search"
s["Type"]="Card"
s["rank"]=s.index
s["rank"]=s["rank"]+1
s.drop(["rank bucket","Price","Product Rating","rank order"],axis=1,inplace=True)
s.to_csv("Xiaomi Visibilty search amazon output.csv",index=False)
#C:\Users\HP\Downloads\brand.csv
#C:\Users\HP\Downloads\Sample data visibility 12_10_22 - papun_test.csv
```
